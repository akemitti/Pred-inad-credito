```python
import os
import zipfile
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import io
from pypdf import PdfReader
import requests
from datetime import datetime
import re
```

```python
def ler_pdf(pdf):
    try:
        reader = PdfReader(pdf)
        textos = []
        for page in reader.pages:
            textos.append(page.extract_text() or "")
        return textos
    except Exception as e:
        print(f"  ! Erro ao ler o stream do PDF: {e}")
        return [f"Erro ao processar PDF: {e}"]
    
def extrair_arquivo(url):
    resultados_extracao = {}

    print(f"Baixando arquivo de: {url}...")
    try:
        response = requests.get(url)
        response.raise_for_status()
        print("Download concluído com sucesso.")
        zip = io.BytesIO(response.content)

        with zipfile.ZipFile(zip, 'r') as zf:
            for nome_arquivo in zf.namelist():
                if nome_arquivo.endswith('.pdf') and not nome_arquivo.startswith("__MACOSX"):

                    print(f"\nProcessando PDF: {nome_arquivo}")

                    with zf.open(nome_arquivo) as pdf_stream:
                        paginas = ler_pdf(pdf_stream)

                        resultados_extracao[nome_arquivo] = paginas

        return resultados_extracao

    except requests.exceptions.RequestException as e:
        print(f"Erro ao baixar o arquivo ZIP: {e}")
        return None
```

```python
url_atas = "https://raw.githubusercontent.com/akemitti/Pred-inad-credito/main/COPOM.zip"

atas = extrair_arquivo(url_atas)
if atas:
    print("\n\n--- Resultados da extração (por página) ---")

    for nome_arquivo, lista_paginas in atas.items():
        print(f"\n==========================================")
        print(f"ARQUIVO: {nome_arquivo} ({len(lista_paginas)} páginas)")
        print(f"==========================================")
```

```python
url_relatorios = "https://raw.githubusercontent.com/akemitti/Pred-inad-credito/main/Estat%C3%ADsticas%20monet%C3%A1rias%20e%20de%20cr%C3%A9dito.zip"

relatorios = extrair_arquivo(url_relatorios)
if relatorios:
    print("\n\n--- Resultados da extração (por página) ---")

    for nome_arquivo, lista_paginas in relatorios.items():
        print(f"\n==========================================")
        print(f"ARQUIVO: {nome_arquivo} ({len(lista_paginas)} páginas)")
        print(f"==========================================")
```

```python
def extrair_data_do_nome(nome_arquivo, tipo='copom'):
    nome_base = os.path.basename(nome_arquivo)

    if tipo == 'copom':
        match = re.search(r'not(\d{8})', nome_base)
        if match:
            data_str = match.group(1)
            return datetime.strptime(data_str, '%Y%m%d')
    else:
        match = re.search(r'(\d{6})', nome_base)
        if match:
            data_str = match.group(1)
            return datetime.strptime(data_str + '01', '%Y%m%d')

    return None


def carregar_documentos(pasta_base, tipo='copom'):
    documentos = []

    if not pasta_base:
        return pd.DataFrame(documentos)

    for arquivo, lista_paginas in pasta_base.items():
        texto = ' '.join([pagina for pagina in lista_paginas if pagina]).strip()
        data = extrair_data_do_nome(arquivo, tipo)

        if texto and data:
            documentos.append({
                'data': data,
                'texto': texto,
                'arquivo': os.path.basename(arquivo),
                'tipo': tipo
            })

    return pd.DataFrame(documentos)
```

```python
df_copom = carregar_documentos(atas, tipo='copom')
print(f"Total de documentos COPOM: {len(df_copom)}")

df_estatisticas = carregar_documentos(relatorios, tipo='estatisticas')
print(f"Total de documentos de Estatísticas: {len(df_estatisticas)}")

df_documentos = pd.concat([df_copom, df_estatisticas], ignore_index=True)
df_documentos = df_documentos.sort_values('data').reset_index(drop=True)

print(f"\nTotal de documentos: {len(df_documentos)}")
print(f"Período: {df_documentos['data'].min().strftime('%Y-%m-%d')} a {df_documentos['data'].max().strftime('%Y-%m-%d')}")

df_documentos.head()
```
